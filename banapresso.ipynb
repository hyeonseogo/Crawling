```python
import pandas as pd
import folium
```

```python
store = pd.read_csv('/content/drive/MyDrive/AI활용 소프트웨어 개발/11. 데이터 분석/data/banapresso_store_with_coords.csv', low_memory=False).dropna().reset_index(drop=True)
store
```

```python
data = store

store_map = folium.Map(location=[data['latitude'].mean(), data['longitude'].mean()], zoom_start=12)

for i in data.index:
    name = '🍧 ' + data.loc[i, 'store_name'] + ' - ' + data.loc[i, 'address']
    popup = folium.Popup(name, max_width=500)
    folium.Marker(location=[data.loc[i, 'latitude'], data.loc[i, 'longitude']],
                  popup=popup, icon=folium.Icon(icon='heart', color='pink')).add_to(store_map)

store_map
```

```python
store_map.save('banapresso_map.html')
```
